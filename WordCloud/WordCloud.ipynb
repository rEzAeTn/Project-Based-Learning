```python
# import Package in program

# import WordCloud
from wordcloud import WordCloud

# import matplotlib
import matplotlib.pyplot as plt

# magik command for jupyter lab
%matplotlib inline
```

```python
# open file and create Data

with open('src/data/movies.txt') as movies:
    text = movies.read()
```

```python
# Create Object from WordCloud

wordcloud = WordCloud(
    width=1000,
    height=500,
    background_color='blue',
    max_font_size=200,
    max_words=50,
    stopwords=None,
    ).generate(text)


# Or use **kwargs Instead of argument

# def run(self, output_path, **kwargs):
#         # Create Object from WordCloud
#         wordcloud = WordCloud(
#         **kwargs,
#         ).generate(self.text)
```

```python
# Data Visualization from matplotlib

plt.imshow(wordcloud)
plt.axis('off')
```

```python
# save output as image

wordcloud.to_file('new.png')
```
