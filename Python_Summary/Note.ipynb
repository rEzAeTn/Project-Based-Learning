# Inheritance in Python
- Inheritance is a fundamental concept in object-oriented programming (OOP) that allows for the creation of a new class that inherits the attributes and methods of an existing class.
- The class from which other classes inherit is called the base class or the parent class, and the class that inherits from the parent class is called the child class or the derived class.\
  **Parent  & AttributsClass** or **Base Class**    --*inherits*-->    **Child Class** or **Derived Class**
```python
class Parent:
    pass

class Child(Parent):    
    pass 
```

### Adding and Overriding Methods & Attributs
inheritance allows a subclass (Child Class) to inherit Methods and Attributes from a base class.\
it also enables the subclass to define its own methods or override inherited ones.\
Child Class can **Extended** (*Add New*) **Attributes** and **Methods** or **Override** (*Modify*) existing ones inherited from the Parent Class.

### Single Inheritance

```python
class ParentClass:
```

***Key Benefits:***

- **Simplicity**
- **Extensibility**

  
### Multilevel Inheritance

**Multilevel inheritance refers to a scenario where a class inherits from a parent class, which in turn inherits from another parent class, creating a chain of inheritance.**\
This type enables the creation of a class hierarchy.

```python
class GrandparentClass:
    # Base class methods and properties

class ParentClass(GrandparentClass):
    # Inherits from GrandparentClass,
    # additional methods and properties

class ChildClass(ParentClass):
    # Inherits from ParentClass,
    # more additional methods or overriding properties
```

***Key Benefits:***

**Definition:** A concept where the suitability of an object for a particular operation is determined by the presence of certain methods and properties, not the object's actual type.

**Purpose:** Emphasizes behavior over type, allowing objects to be used interchangeably if they have the necessary methods or properties.

## Encapsulation
Encapsulation in Python is the bundling of data (attributes) and methods (function) that operate on the data into a single unit, or class, while controlling access to the internal state to protect it from unintended modifications.\
Encapsulation enhances **Data integrity**, **Security**, and **Abstraction**.

- **Public Members** are accessible from any part of the code.
- **Protected Members** are accessible within the class and by derived classes (subclasses).
- **Private Members** are accessible only within the class defining them.

#### Roles of Encapsulation
- **Data Hiding:** Prevents external code from accessing internal object state directly.

- **Data Protection:** Guards data against unintentional corruption.

- **Abstraction:** Hides complex implementation details from the user.

### Encapsulation in Python(object-oriented programming (OOP))
- ***Public Attributes/Methods:*** By default, all members of a Python class are public, meaning they can be accessed from anywhere in the code.
- ***Protected Attributes/Methods:*** Indicated by a single underscore prefix _, these are intended for internal use by the class and its subclasses. This is more of a convention than a strict rule and serves as a hint to other developers.
- ***Private Attributes/Methods:*** Denoted by a double underscore prefix __, these members are name-mangled to make them harder (but not impossible) to access from outside the class. This mechanism doesn't fully prevent access from outside but acts as a stronger suggestion that such members are intended for internal use only.

## Overriding `__repr__` for Unambiguous Representations

- Override the `__repr__` method in Python To provide an unambiguous representation of an object.
- `__repr__` method return a string that could be used to recreate the object.
- `__repr__` Method: This method returns a detailed and unambiguous string representation of the object.
- Usage: The repr() function calls `__repr__` method.

```python
class CustomClass:
    def __init__(self, attribute_1, attribute_2):
        self.attribute_1 = attribute_1  # Initialize Attribute
        self.attribute_2 = attribute_2  # Initialize Attribute

    def __repr__(self):
        # Return a detailed string representation of the object
        # This string should be unambiguous and, if possible, mimic the call used to create the object
        return f"{self.__class__.__name__}(attribute_1='{self.attribute_1}, attribute_2='{self.attribute_2}, ')"

# Create an object of CustomClass
obj = CustomClass("example_value")

```



## Differences between `__str_`_ and `__repr__`

#### Purpose:

`__str__`: For readable and human-friendly representation of an object, for end-users output.

`__repr__`: For unambiguous and detailed representation, developer-focused output ( a string that can be used to recreate the object )


#### Usage:

`__str__`: Used by `print()` and `str()`.\
When you print an object, Python calls the `__str__` method.

`__repr__`: Used by `repr()`, and in interactive interpreter sessions (like *Jupyter notebooks*).\
When you type an object in a cell and execute it, the `__repr__` method is called.


#### Differences in Applications:

`__str__`: Enhances readability for general display.

`__repr__`: Enhances debugging and logging by providing detailed object information.


#### Example
##### In short:
`__str__` is for readability, making objects human-friendly.\
`__repr__` is for accuracy, providing details developers need for debugging.

```python

        self.attribute = attribute

        return f"CustomClass with attribute: {self.attribute}"  # Readable representation for end-users

# __str__ is called when using print()
```

## Logical Operators in Python (`and`, `or`, `not`)
### `and`:
#### Concept:
Similar to multiplication(`*`).

#### Evaluation:
If the first expression is falsy, it returns the first expression(False).

If the first expression is truthy, it evaluates and returns the second expression(True or False).

#### Example:
```python
False and True  # Returns False
True and 42     # Returns 42
```
    
### `or:`
#### Concept:
Similar to addition(`+`).

#### Evaluation:
If the first expression is truthy, it returns the first expression.
If the first expression is falsy, it evaluates and returns the second expression.

#### Example:
```python
True or False   # Returns True
False or 42     # Returns 42
```
    
### `not`:
#### Concept:
Logical negation.

#### Evaluation:
Returns the opposite boolean value of the expression.

#### Example:
```python
not True   # Returns False
not False  # Returns True
```

## `if __name__ == "__main__":`

### `__name__`:
Built-in variable indicating the script's execution context.

### `__main__`:
When the script runs directly, __name__ is set to "__main__".

### `if __name__ == "__main__":`:
Ensures the code within the block runs only when the script is executed directly.

### Usage:
To run specific code only when the script is run directly, not when imported.

### Condition:
**True:** When the script is executed directly.
**False:** When the script is imported as a module.

### Example:
```python
def main():
    print("This runs only if the script is executed directly.")

if __name__ == "__main__"
    main()
```

## Python Path
In Python, you can manage the environment variable `PYTHONPATH` to ensure your custom modules and packages are recognized by the interpreter.
Useful for managing your development environment and ensuring your scripts can find all the necessary modules.

```python
import sys

# Check current system paths
sys.path

# Output:
['/home/reza/Project-Based-Learning/Python_Summary', '/usr/lib/python3.9', '/usr/lib/python3.9/lib-dynload', '...', '...']
```

### Import Package
To import custom packages in your projects, start all imports from `src` directory.
```python
from src.utils.package_name import module
```

### Add Path to `sys.path`
You can dynamically add paths to the `sys.path` list to ensure your modules are found.

#### Example in Python
```python
import sys
sys.path.append('/path/to/your/module')

# Now you can import your custom module
import custom_module
```

### Linux Commands
Add Path to `sys.path`
Set the `PYTHONPATH` environment variable to include the current working directory(`pwd`).

```bash
# Add current directory to PYTHONPATH
export PYTHONPATH=$PYTHONPATH:$(pwd)

# Verify the current PYTHONPATH
echo $PYTHONPATH

# Output: /existing/paths:/home/user/current_directory
```
