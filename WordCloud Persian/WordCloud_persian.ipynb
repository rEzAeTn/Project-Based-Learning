```python
# import Package in program

# import WordCloud
from wordcloud import WordCloud

# import matplotlib
import matplotlib.pyplot as plt

# magik command for jupyter lab
%matplotlib inline
```

```python
# open file and create Data
with open('Data/text.txt', encoding='utf-8') as txt:
    read_text = txt.read()
```

```python
# Filter Persian Text

def filter_persian(text: str) -> str:
    text = text.replace('\u200c', ' ')
    persian_alphabet = 'ابپتثجچحخدذرزژسشصضطظعغفقکگلمنوهی '
    filtered_text = ''.join([char for char in text if char in persian_alphabet])
    return filtered_text


# Create Object from filter_persian()

filtered_text = filter_persian(read_text)
```

```python
# Modification of Persian texts

import arabic_reshaper
from bidi.algorithm import get_display


reshaped_text = arabic_reshaper.reshape(filtered_text)
text = get_display(reshaped_text)
```

```python
# Set the font path to a Persian font
persian_font_path = 'Font/font.ttf'

# create stopwords
with open('stopwords.txt', encoding='utf-8') as stopwords:
    persian_stopwords = stopwords.read().replace('\u200c', ' ')
    persian_stopwords = set(persian_stopwords.split('\n'))


print(persian_stopwords)
```

```python
# Create Object from WordCloud

wordcloud = WordCloud(
    font_path=persian_font_path,
    width=1000,
    height=500,
    background_color='white',
    max_font_size=200,
    max_words=50,
    min_word_length=0,
    stopwords=persian_stopwords,
    ).generate(text)


# Or use **kwargs Instead of argument

# def run(self, output_path, **kwargs):
#         # Create Object from WordCloud
#         wordcloud = WordCloud(
#         **kwargs,
#         ).generate(self.text)


```

```python
# Data Visualization from matplotlib

plt.imshow(wordcloud)
plt.axis('off')
```
